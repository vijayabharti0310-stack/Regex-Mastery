# RegEx (Regular Expression)

```python
import re
```

```python
#match only digits
text="Order number 12345 arrived on 25-11-10"       #r-,
print(re.findall(r'\d+',text))
```

```python
text="Order number 12345 arrived on 25-11-10"      # extract value one by one
print(re.findall(r'\d',text))
```

```python
#Match words starting with capital letter
text="Python Regex Is Powerful.the python Iron"   # + -> will look same pattern 
print(re.findall(r'\b[A-Z][a-z]+\b',text))
```

```python
text="Python Regex Is Powerful.the python Iron"   # + -> will look same pattern 
print(re.findall(r'\b[a-z][a-z]+\b',text))    # this patern says [a-z][a-z]--- word starts with small letter and also ends with small letter.
```

```python
#find words ending with "ing"
text="I am learning coding and running daily"
#print(re.findall(r'\b[a-z]['i-ng']+\b',text))
print(re.findall(r'\b\w+ing\b',text))
```

```python
#find all secial characters
text="Email me @ test#regex.com!"
print(re.findall(r'[^A-Za-z0-9\s]',text))
```

```python
# validate a mobile number (10 digits)
pattern=r'^\d{10}$'
numbers=["9876543210","98765","9876543210"]

for num in numbers:
    print(num,bool(re.match(pattern,num)))
```

```python
#postal code (6 digits)
pattern=r'^\d{6}$'
codes=['123456','6543','654987','908076','87895']
for i in codes:
    print(i,bool(re.match(pattern,i)))
```

```python
#Aadhar (12 digits with 2 spaces)
pattern=r'^\d\s{12}$'
numbers=['1234 5678 9876','11111 2345 5678']
for i in numbers:
    print(i,bool(re.match(pattern,i)))       # aadhar sholud not be started with zero 0 not 1.
```

```python
# Extract domain from email
emails=['abc@gmail.com','john@yahoo.com','test@outlook.in','test2@outlook.in']
for email in emails:
    match=re.search(r'@([A-Za-z0-9.-]+)',email)     # capital A to Z to small a to z , 0 to 9 with . and - also.
    if match:
        print(match)
        print(email,"->",match.group(0))
```

```python
# Using groups to extract names
text="Name: John Doe, Age: 29"
match=re.search(r'Name:\s([A-Za-z]+)\s([A-Za-z]+),\sAge:\s(\d+)',text)
if match:
    print(match)
    print("First Name:",match.group(1))
    print("Last Name:",match.group(2))
    print("Age:",match.group(3))
    
```

```python
#Validate password (8+ character, at least 1 uppercase,1 digit)
pattern= r'^(?=.*[A-Z])(?=.*\d).{8,}$'
passwords=['Pass1234','password','PASS14@']

for p in passwords:
    print(p,bool()re.match(pattern,p))
```

```python
# Match repeated words (using backreferences)
text="This is is a test test line"
print(re.findall(r'\b(\w+)\s+\1\b',text))
```

```python
#Extract quoted text
text='He said, "Hello there!1" and she replied,"Hi!"'
print(re.findall(r'"(.*?)"',text))
```

```python
# Find all uppercase word (acronyms)
text="NASA and ISRO are space agencies"
print(re.findall(r'\b[A-Z]{2,}\b',text))
```

```python
# Check if a string is a valid hex color
colors = ["#FFF", "#1a2b3c", "#GGGGGG"]
pattern = r'^#([A-Fa-f0-9]{3}|[A-Fa-f0-9]{6})$'

for color in colors:
    print(color, bool(re.match(pattern, color)))
```
