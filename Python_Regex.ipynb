# Regular Expressions

* A regular expression is a sequence of characters that uses a search pattern to find a string or set of strings ie., ‘find and replace’ –like operations.

* It can detect the presence or absence of a text by matching it with a particular pattern and splitting a pattern into one or more sub-patterns.

* Python provides a re-module that supports the use of regex in Python.

* Its primary function is to offer a search that takes a regular expression and a string. Here, it either returns the first match or else none.

* It is so powerful that it can extract a particular information from a text. So we can make our own Web Crawlers and scrappers in python.


```python
import re
```

```python
#Match only digits
text = "Order number 12345 arrived on 2025-11-10"
print(re.findall(r'\d+', text))
```

```python
#Match words starting with a capital letter
text = "Python Regex Is Powerful.the python Iron"
#print(re.findall(r'\b[A-Z][a-z]+\b', text))

#print(re.findall(r'\b[a-z]+\b', text))
```

```python
#Find words ending with "ing"
text = "I am learning coding and running daily"
print(re.findall(r'\b\w+ing\b', text))

```

```python
# Find all special characters
text = "Email me @ test#regex.com!"
print(re.findall(r'[^A-Za-z0-9\s]', text))
```

```python
# Validate a mobile number (10 digits)
pattern = r'^\d{10}$'
numbers = ["9876543210", "98765", "9876543210"]

for num in numbers:
    print(num, bool(re.match(pattern, num)))

```

```python
#Extract domain from email
emails = ["abc@gmail.com", "john@yahoo.com", "test@outlook.in",'test2@outlook.in']
for email in emails:
    match = re.search(r'@([A-Za-z0-9.-]+)', email)
    if match:
        print(match)
        print(email, "→", match.group(0))

```

```python
#Replace all digits with X
text = "My OTP is 987654"
print(re.sub(r'\d', 'X', text))
```

```python
#Extract all words
text = "Python 3.12 released in 2025!"
print(re.findall(r'\b[A-Za-z]+\b', text))

```

```python
# Split string by punctuation
text = "apple, banana; orange. mango"
print(re.split(r'[;,.]', text))

```

```python
# Extract hashtags and mention

text = "Follow @python_dev and learn from #Re1gex #Tutorial"
print("Mentions:", re.findall(r'@\w+', text))
print("Hashtags:", re.findall(r'#\w+', text))

```

```python
# Match dates (dd/mm/yyyy or dd-mm-yyyy)
text = "Today's date is 13/11/2025, tomorrow is 14-11-2025."
print(re.findall(r'\b\d{2}[-/]\d{2}[-/]\d{4}\b', text))

```

```python
# Extract IP addresses
text = "Server1: 19.168.1.1, Server2: 10.0.0.255"
print(re.findall(r'\b\d{1,3}(?:\.\d{1,3}){3}\b', text))

```

```python
# Extract URLs
text = "Visit https://test.com or http://example.org for details"
print(re.findall(r'https?://[^\s]+', text))

```

```python
# Extract words of specific length (e.g., 4-letter words)
text = "This text has some four word test data but after this there is nothing"
print(re.findall(r'\b[a-zA-Z]{4}\b', text))

```

```python
# Using groups to extract names
text = "Name: John Doe, Age: 29"
match = re.search(r'Name:\s([A-Za-z]+)\s([A-Za-z]+),\sAge:\s(\d+)', text)
if match:
    print(match)
    print("First Name:", match.group(1))
    print("Last Name:", match.group(2))
    print("Age:", match.group(3))
```

```python
# Validate password (8+ chars, at least 1 uppercase, 1 digit)
pattern = r'^(?=.*[A-Z])(?=.*\d).{8,}$'
passwords = ["Pass1234", "password", "PASS14@"]

for p in passwords:
    print(p, bool(re.match(pattern, p)))

```

```python
# Match repeated words (using backreference)
text = "This is is a test test line"
print(re.findall(r'\b(\w+)\s+\1\b', text))

```

```python
# Extract quoted text
text = 'He said, "Hello there!1" and she replied, "Hi!"'
print(re.findall(r'"(.*?)"', text))
```

```python
#Find all uppercase words (acronyms)
text = "NASA and ISRO are space agencies"
print(re.findall(r'\b[A-Z]{2,}\b', text))
```

```python
# Check if a string is a valid hex color
colors = ["#FFF", "#1a2b3c", "#GGGGGG"]
pattern = r'^#([A-Fa-f0-9]{3}|[A-Fa-f0-9]{6})$'

for color in colors:
    print(color, bool(re.match(pattern, color)))

```
